```python
loader = PyPDFDirectoryLoader(path="pdf_files/",glob="**/*.pdf")
loader
```

```python
pdfs = loader.load_and_split()
pdfs
```

```python
len(pdfs)
```

```python
for x in pdfs:
    print("Page Source:",x.metadata['source'])
    print("Page No:",x.metadata['page'])
    print("Content:",x.page_content)
    print('----'*30)
```
